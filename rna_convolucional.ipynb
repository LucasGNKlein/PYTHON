```python
# ==========================================
# Exemplo Completo de CNN - CIFAR-10
# Autor: Prof. Marcos (base didática)
# ==========================================

# Importação das bibliotecas principais
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical

# -----------------------------------------------------------
# 1️⃣ Carregar e preparar o conjunto de dados
# -----------------------------------------------------------
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normaliza os valores dos pixels para [0,1]
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Converte rótulos para one-hot encoding
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)

# Lista das classes do CIFAR-10
classes = ['Avião', 'Automóvel', 'Pássaro', 'Gato', 'Cervo',
           'Cachorro', 'Sapo', 'Cavalo', 'Navio', 'Caminhão']

# Exibe algumas imagens do dataset
plt.figure(figsize=(10,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(x_train[i])
    plt.title(classes[int(y_train[i])])
    plt.axis('off')
plt.suptitle('Exemplos de imagens do CIFAR-10')
plt.show()

# -----------------------------------------------------------
# 2️⃣ Construção da Rede Neural Convolucional (CNN)
# -----------------------------------------------------------
model = Sequential()

# Primeira camada convolucional + pooling
model.add(Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)))
model.add(MaxPooling2D((2,2)))

# Segunda camada convolucional + pooling
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPooling2D((2,2)))

# Terceira camada convolucional
model.add(Conv2D(128, (3,3), activation='relu'))

# Flatten para transformar os mapas de características em vetor
model.add(Flatten())

# Camadas densas totalmente conectadas
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))  # evita overfitting
model.add(Dense(10, activation='softmax'))  # saída com 10 classes

# Exibe o resumo da rede
model.summary()

# -----------------------------------------------------------
# 3️⃣ Compilação e treinamento do modelo
# -----------------------------------------------------------
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Treinamento (ajuste de pesos)
history = model.fit(x_train, y_train_cat,
                    epochs=10,
                    batch_size=64,
                    validation_data=(x_test, y_test_cat))

# -----------------------------------------------------------
# 4️⃣ Avaliação do modelo
# -----------------------------------------------------------
loss, acc = model.evaluate(x_test, y_test_cat, verbose=0)
print(f'Acurácia no conjunto de teste: {acc*100:.2f}%')

# Gráfico de desempenho
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Treino')
plt.plot(history.history['val_accuracy'], label='Validação')
plt.title('Acurácia durante o treinamento')
plt.xlabel('Época')
plt.ylabel('Acurácia')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Treino')
plt.plot(history.history['val_loss'], label='Validação')
plt.title('Perda durante o treinamento')
plt.xlabel('Época')
plt.ylabel('Loss')
plt.legend()
plt.show()

# -----------------------------------------------------------
# 5️⃣ Testando previsões
# -----------------------------------------------------------
# Seleciona 10 imagens aleatórias do conjunto de teste
indices = np.random.randint(0, len(x_test), 10)
sample_images = x_test[indices]
sample_labels = y_test[indices]

# Faz previsões
predictions = model.predict(sample_images)
predicted_classes = np.argmax(predictions, axis=1)

# Exibe previsões vs. rótulos reais
plt.figure(figsize=(10,4))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(sample_images[i])
    true_label = classes[int(sample_labels[i])]
    pred_label = classes[predicted_classes[i]]
    color = 'green' if true_label == pred_label else 'red'
    plt.title(f"Previsto: {pred_label}\nVerdadeiro: {true_label}", color=color, fontsize=9)
    plt.axis('off')
plt.suptitle('Resultados das Previsões da CNN')
plt.show()
```
